```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('data.csv')
```

```python
# Step 3: Get basic info about dataset
data.info()  # shows number of rows, columns, null values, and data types
```

```python
# Step 4: Check for missing values in each column
data.isnull().sum()  # useful to detect if any data is missing
```

```python
# Drop the unnamed column if it exists
data.drop(columns=["Unnamed: 32"], axis=1, inplace=True, errors='ignore')
```

```python
data.head()
```

```python
# Step 6: Check how many benign and malignant cases are there
data['diagnosis'].value_counts()  # counts M and B in diagnosis column
```

```python
# Plot the count of each diagnosis type
sns.countplot(x='diagnosis', data=data)
plt.title("Diagnosis Count (B = Benign, M = Malignant)")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
# Example usage
le = LabelEncoder()  # create object
data['diagnosis'] = le.fit_transform(data['diagnosis'])  # apply on column

```

```python
data.head()
```

```python
# Step 8: View statistical summary of dataset
data.describe()  # shows mean, min, max, std deviation etc.
```

```python
# Step 9: Create a heatmap to see how features relate to each other
plt.figure(figsize=(12, 10))  # set figure size
sns.heatmap(data.corr(), cmap='coolwarm')  # show correlation between features
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# Step 11: Boxplot to compare feature values between benign and malignant
# Helps to see differences and outliers

plt.figure(figsize=(6,4))
sns.boxplot(x='diagnosis', y='radius_mean', data=data)
plt.title("Radius Mean by Diagnosis")
plt.show()
```

```python
# Feature Selection and Scaling
```

```python
X = data.drop('diagnosis', axis=1)
y = data['diagnosis']

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
##Train-Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))

```
